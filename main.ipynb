# Предобработка датафреймов

Пока что все данные в датафреймах - строковые величины, хотя отгрузка по смыслу должна быть числом (где так-же присутствуют и строковые значения типа в ноль и т.д., их тоже нужно обрабатывать), месяц - датой (yyyy-mm)

Преобразуем датафрейм

Судя по таблице prolongation для 1 проекта может быть несколько менеджеров, сменяющихся во времени, для корректного подсчета коэф-та это нужно учесть - учитывая конечную дату проекта для данного менеджера: делать null все значения после даты завершения для данного менеджера и перед датой завершения для предыдущего по времени менеджера для данного проекта если он есть

Необходимо проверять для каждого проекта конечную дату, если в значениях отгрузок есть стоп или end в конечный месяц или ранее во времени то этот проект просто исключается

Грамотно обработать значения отгрузок - если мы имеем null то он должен остаться и не применяться в дальнейших вычислениях(или хотябы не изменяться на другое значение типа 0, т.к. это ломает логику вычисления коэф-та), потому что по смыслу в месяцах где отгрузка null не происходит никаких действий. соответственно для значений стоп и end находящихся после конечного месяца проекта во времени (если такие есть) используем null. если же значение в ноль, т.е. 0 (уже с учетом агрегации по дублированным проектам)- при вычислении коэф-та пролонгации используем предыдущий месяц как и сказано в ТЗ

## Преобразуем месяцы

для корректного сравнения дат

```python
MONTHS = {
    'январь':1,'февраль':2,'март':3,'апрель':4,'май':5,'июнь':6,
    'июль':7,'август':8,'сентябрь':9,'октябрь':10,'ноябрь':11,'декабрь':12
}

def parse_month_year(s):
    """
        :param s: строка вида "Май 2023" или "май 2023"
        :return: datetime или None если s некоректен
    """
    if pd.isna(s):
        return None
    s = str(s).strip().lower()
    parts = s.split()
    if len(parts) < 2:
        return None
    mon = parts[0]
    year = int(parts[1])
    m = MONTHS.get(mon)
    if not m:
        return None
    return datetime(year, m, 1)
```

## Обрабатываем значения отгрузки

```python
def class_and_parse(x):
    """
    Прверяет строковые значения отгрузки
    :param x: строка с возможными типами стоп, end, в ноль и вещ. число
    :return: {тип (null, stop, zero, numeric, unknow), значение (NaN, 0, число)}
    """
    if pd.isna(x) or str(x).strip() == '':
        return {'kind':'null', 'num':np.nan}
    s = str(x).strip().lower().replace('\xa0',' ').replace(' ',' ')
    if s in ['стоп','end']:
        return {'kind':'stop', 'num':np.nan}
    if 'в ноль' in s or s=='в ноль' or s=='0' or s=='0,00' or s=='0.00':
        return {'kind':'zero', 'num':0.0}
    # теперь парсим вещ. число
    s2 = s.replace(' ','').replace('\xa0','').replace(',','.')
    try:
        v = float(s2)
        return {'kind':'numeric', 'num':v}
    except:
        return {'kind':'unknown', 'num':np.nan}

parsed = fin_long['raw_value'].apply(class_and_parse)
# записываем
fin_long['kind'] = parsed.apply(lambda d: d['kind'])
fin_long['num'] = parsed.apply(lambda d: d['num'])
```

## Агрегация дублей

```python
agg = fin_long.groupby(['id','month_dt']).agg(
    sum_amount = ('num', lambda s: np.nansum([v for v,k in zip(s, fin_long.loc[s.index,'kind']) if k=='numeric'])),    # вычисляем сумму num только для kind = numeric
    any_numeric = ('kind', lambda s: any(k=='numeric' for k in s)),    # true если есть хотя-бы 1 строка с numeric в этой группе
    any_nonnull = ('kind', lambda s: any(k!='null' for k in s)),    # True если хотя-бы 1 значение не null
    any_zero_marker = ('kind', lambda s: any(k=='zero' for k in s)),    # True если есть хотя-бы 1 значение 0
).reset_index()
```

```python
def calc_all_zero(row):
    """
    Флаг all_parts_zero для перехода к след. месяцу: нет ни одной numeric > 0, и есть хотя бы один non-null
    (т.е. были записи), и все числовые части равны 0 (или отсутствуют)
    :param row: строка датафрейма
    :return: является ли нулевой
    """
    # any_numeric True, но sum_amount == 0 – numeric есть но = 0
    if (not row['any_nonnull']):
        return False
    if row['any_numeric'] and row['sum_amount'] == 0:
        return True
    # если не было numeric, но были 'zero' метки – считаем как all zero
    if (not row['any_numeric']) and row['any_zero_marker']:
        return True
    return False

agg['all_parts_zero'] = agg.apply(calc_all_zero, axis=1)
```

Для каждого id: берем строки prolong (AM_from_prolong + prolong_month_dt), сортируем по prolong_month_dt, и считаем interval

```python
# для каждого проекта сортируем по дате
prolong = prolong.sort_values(['id','prolong_month_dt'])
```

```python
mgr_intervals = []

for pid, g in prolong.groupby('id'):
    g_sorted = g.sort_values('prolong_month_dt')
    prev_end = None
    for _, row in g_sorted.iterrows():
        this_end = row['prolong_month_dt']
        this_mgr = row['AM_from_prolong']
        start = (prev_end + relativedelta(months=1)) if prev_end is not None else datetime(1900,1,1)
        end = this_end
        mgr_intervals.append({'id':pid, 'manager':this_mgr, 'start':start, 'end':end})
        prev_end = this_end

mgr_intervals = pd.DataFrame(mgr_intervals)
```

```python
# берём все комбинации (id, month_dt) из agg и присваиваем менеджера, чей interval покрывает month_dt
agg_with_mgr = agg.merge(mgr_intervals, on='id', how='left')
```

```python
agg_with_mgr.head(15)
```

```python
def assign_mgr(row):
    """
    Присваивает менеджера данной записи, если его interval покрывает month_dt
    :param row: df series
    :return: менеджер
    """
    if pd.isna(row['start']) or pd.isna(row['end']) or pd.isna(row['month_dt']):
        return np.nan
    if (row['month_dt'] >= row['start']) and (row['month_dt'] <= row['end']):
        return row['manager']
    return np.nan

agg_with_mgr['assigned_manager'] = agg_with_mgr.apply(assign_mgr, axis=1)
```

## Считаем коэффициенты k1 k2

```python
full2 = full.copy()

# проект считается завершённым в project_final_month
proj_final_map = proj_final.set_index("id")["project_final_month"].to_dict()
full2["final_month"] = full2["id"].map(proj_final_map)

# признак последнего месяца проекта
full2["is_last_month"] = full2["month_dt"] == full2["final_month"]
```

```python
# суммы по месяцам, удобно в pivot
pivot = full2.pivot_table(
    index="id",
    columns="month_dt",
    values="sum_amount"
)

# менеджеры по проектам
mgr_map = (
    full2.dropna(subset=["assigned_manager"])
        .drop_duplicates(subset=["id"])
        .set_index("id")["assigned_manager"]
        .to_dict()
)
```

Функции расчета k1 и k2

```python
def calc_k1(target_month):
    prev_month = target_month - relativedelta(months=1)

    # проекты, завершившиеся в prev_month
    ended = full2.loc[
        (full2["month_dt"] == prev_month) &
        (full2["is_last_month"])
    ]["id"].unique()

    if len(ended) == 0:
        return pd.DataFrame()

    df = pd.DataFrame({"id": ended})
    df["manager"] = df["id"].map(mgr_map)
    df["last"] = df["id"].map(lambda pid: pivot.loc[pid, prev_month])
    df["next"] = df["id"].map(lambda pid: pivot.loc[pid, target_month])

    df = df.fillna(0)

    result = df.groupby("manager").agg(
        k1_num=("next", "sum"),
        k1_den=("last", "sum")
    )
    result["k1"] = result["k1_num"] / result["k1_den"]
    result["month"] = target_month
    return result.reset_index()
```

```python
def calc_k2(target_month):
    m1 = target_month - relativedelta(months=1)
    m2 = target_month - relativedelta(months=2)

    # проекты, завершившиеся в m2
    ended = full2.loc[
        (full2["month_dt"] == m2) &
        (full2["is_last_month"])
    ]["id"].unique()

    if len(ended) == 0:
        # создаём пустой df с нужными колонками
        return pd.DataFrame(columns=["manager","month","k2"])

    df = pd.DataFrame({"id": ended})
    df["manager"] = df["id"].map(mgr_map)
    df["last"] = df["id"].map(lambda pid: pivot.loc[pid, m2])
    df["m1"] = df["id"].map(lambda pid: pivot.loc[pid, m1])
    df["m2"] = df["id"].map(lambda pid: pivot.loc[pid, target_month])

    # условие для k2 — НЕТ пролонгации на m1, но есть на M
    df = df[(df["m1"] == 0) & (df["m2"] > 0)]

    df = df.fillna(0)

    if df.empty:
        # пустой df с нужными колонками
        return pd.DataFrame(columns=["manager","month","k2"])

    result = df.groupby("manager").agg(
        k2_num=("m2", "sum"),
        k2_den=("last", "sum")
    )
    result["k2"] = result["k2_num"] / result["k2_den"]
    result["month"] = target_month
    return result.reset_index()[["manager","month","k2"]]
```

## Формируем итоговую таблицу

```python
all_results = []
all_managers = sorted(full2['assigned_manager'].dropna().unique())

for M in months_dt:
    k1_res = calc_k1(M)
    k2_res = calc_k2(M)

    if k1_res is None or not set(["manager","k1"]).issubset(k1_res.columns):
        k1_res = pd.DataFrame(columns=["manager","k1"])

    if k2_res is None or not set(["manager","k2"]).issubset(k2_res.columns):
        k2_res = pd.DataFrame(columns=["manager","k2"])

    # создаём dataframe со всеми менеджерами на текущий месяц
    month_df = pd.DataFrame({"manager": all_managers})
    month_df["month"] = M

    # подставляем k1
    month_df = month_df.merge(k1_res[["manager","k1"]], on="manager", how="left")

    # подставляем k2
    month_df = month_df.merge(k2_res[["manager","k2"]], on="manager", how="left")

    all_results.append(month_df)

k_table = pd.concat(all_results, ignore_index=True)
```

Преобразуем в широкий вид, добавим годовые значения и значения для всего отдела

```python
# Копируем, чтобы не менять оригинал
k_table2 = k_table.copy()

# Считаем средние k1/k2 по месяцам для отдела (игнорируем NaN)
dept_row = k_table2.groupby("month").agg(
    k1=("k1", "mean"),
    k2=("k2", "mean")
).reset_index()

dept_row["manager"] = "Отдел"

# добавляем к таблице
k_table2 = pd.concat([k_table2, dept_row], ignore_index=True, sort=False)
```

```python
# сначала pivot для k1
k1_wide = k_table2.pivot(index="manager", columns="month", values="k1")
k2_wide = k_table2.pivot(index="manager", columns="month", values="k2")

# объединяем k1/k2 в одну таблицу
k_wide = pd.DataFrame(index=k1_wide.index)

for month in k1_wide.columns:
    k_wide[f'k1 {month.strftime("%b")}'] = k1_wide[month]
    k_wide[f'k2 {month.strftime("%b")}'] = k2_wide[month]

# добавляем столбцы за год
k_wide["k1 год"] = year_summary.set_index("manager")["k1_year"]
k_wide["k2 год"] = year_summary.set_index("manager")["k2_year"]

# сортируем менеджеров
k_wide = k_wide.sort_index()
```
